```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def my_cdist(X, Y):
    return np.sqrt(np.sum((X[:, np.newaxis, :] - Y[np.newaxis, :, :]) ** 2, axis=2))

def moving_average_vector(point, neighbors):
    # Calculate the mean vector difference between the point and its neighbors
    return np.mean(neighbors - point, axis=0)

# Define a simple 2-layer neural network function
def simple_nn(input_vector, weights1, weights2):
    hidden = np.dot(input_vector, weights1)
    output = np.dot(hidden, weights2)
    return np.mean(output)

# Define a function to calculate the probability of belonging to a cluster
def calculate_probability(distance, threshold):
    return np.exp(-distance / threshold)

# Define a function to calculate local curvature for a sequence of points
def calculate_local_curvature(points):
    curvatures = []
    for i in range(1, len(points) - 1):
        a = points[i - 1]
        b = points[i]
        c = points[i + 1]
        ba = a - b
        bc = c - b
        curvature = np.linalg.norm(np.cross(ba, bc)) / (np.linalg.norm(ba) * np.linalg.norm(bc))
        curvatures.append(curvature)
    return np.array(curvatures)


class ImprovedTemporalSpatialClusterWithNN:
    def __init__(self, density_threshold, temporal_threshold, n_neighbors, angle_threshold):
        self.density_threshold = density_threshold
        self.temporal_threshold = temporal_threshold
        self.n_neighbors = n_neighbors
        self.angle_threshold = angle_threshold
        self.nn_weights1 = np.random.rand(3, 5)  # Initialize random weights for the first layer
        self.nn_weights2 = np.random.rand(5, 1)  # Initialize random weights for the simple NN

    def update_nn_weights(self, curvature):
        # Simulate online learning by adjusting weights based on the curvature
        learning_rate = 0.1
        curvature_mean = np.mean(curvature)

        self.nn_weights1 += learning_rate * curvature_mean
        self.nn_weights2 += learning_rate * curvature_mean

    def fit_predict(self, X):
        if len(X) < 2 * self.n_neighbors + 1:
            raise ValueError("Insufficient data points for the given number of neighbors.")

        # Calculate the distance matrix
        dist_matrix = np.linalg.norm(X[:, np.newaxis] - X, axis=2)

        # Perform spatial clustering based on density
        density = np.sum(dist_matrix < self.density_threshold, axis=1)
        labels = -np.ones(len(X), dtype=int)
        cluster_id = 0
        spatial_clusters = {}

        for i in range(len(X)):
            if density[i] < self.density_threshold or labels[i] != -1:
                continue
            spatial_clusters[cluster_id] = []
            neighbors = np.where(dist_matrix[i] < self.density_threshold)[0]
            for neighbor in neighbors:
                if density[neighbor] >= self.density_threshold:
                    labels[neighbor] = cluster_id
                    spatial_clusters[cluster_id].append(neighbor)
            cluster_id += 1

        line_clusters = {}

        # Improve the clusters temporally with a simple NN and probabilistic approach
        for cluster_id, spatial_cluster in spatial_clusters.items():
            temporal_order = spatial_cluster.copy()
            while temporal_order:
                current = temporal_order.pop(0)
                neighbors = np.where(dist_matrix[current] < self.temporal_threshold)[0]

                # Update the neural network weights based on the calculated curvature
                local_curvature = calculate_local_curvature(X[spatial_clusters[cluster_id]])
                self.update_nn_weights(local_curvature)

                line_score = simple_nn(local_curvature.mean(), self.nn_weights1, self.nn_weights2)
                if line_score > 0.5:
                    line_clusters[cluster_id] = True

                for neighbor in neighbors:
                    if labels[neighbor] == -1:
                        if neighbor > self.n_neighbors and neighbor < len(X) - self.n_neighbors:
                            prev_vector = moving_average_vector(X[neighbor], X[neighbor-self.n_neighbors:neighbor])
                            next_vector = moving_average_vector(X[neighbor], X[neighbor+1:neighbor+self.n_neighbors+1])

                            if np.linalg.norm(prev_vector) == 0 or np.linalg.norm(next_vector) == 0:
                                continue

                            prev_vector = simple_nn(prev_vector, self.nn_weights1, self.nn_weights2)
                            next_vector = simple_nn(next_vector, self.nn_weights1, self.nn_weights2)
                            cosine_angle = np.dot(prev_vector, next_vector) / (np.linalg.norm(prev_vector) * np.linalg.norm(next_vector))

                            prob = calculate_probability(cosine_angle, self.angle_threshold)

                            if cosine_angle > self.angle_threshold and prob > 0.5:
                                labels[neighbor] = cluster_id
                                temporal_order.append(neighbor)
                        else:
                            labels[neighbor] = cluster_id
                            temporal_order.append(neighbor)

        # Merge clusters that are considered lines
        new_labels = labels.copy()
        for cluster_id, is_line in line_clusters.items():
            if is_line:
                members = np.where(labels == cluster_id)[0]
                for member in members:
                    neighbors = np.where(dist_matrix[member] < self.temporal_threshold)[0]
                    neighbor_clusters = [labels[i] for i in neighbors if labels[i] in line_clusters]

                    if len(neighbor_clusters) > len(neighbors) / 3:
                        new_labels[member] = min(neighbor_clusters)

        return new_labels

    def merge_clusters(self, X, labels, merge_threshold=0.5):
        unique_labels = np.unique(labels)
        cluster_centers = {}

        # Calculate the center of each cluster
        for label in unique_labels:
            if label == -1:  # Ignore noise
                continue
            members = X[labels == label]
            center = np.mean(members, axis=0)
            cluster_centers[label] = center

        # Calculate the distance between each pair of cluster centers
        centers_list = np.array(list(cluster_centers.values()))
        dists = my_cdist(centers_list, centers_list)

        # Create a mapping from old labels to new labels (for merging)
        label_mapping = {}
        taken_labels = set()
        for i, label1 in enumerate(unique_labels):
            if label1 in taken_labels or label1 == -1:
                continue
            for j, label2 in enumerate(unique_labels):
                if label2 in taken_labels or label1 == label2 or label2 == -1:
                    continue
                if dists[i, j] < merge_threshold:
                    label_mapping[label2] = label1
                    taken_labels.add(label2)

        # Apply the label mapping to merge clusters
        new_labels = np.array([label_mapping.get(x, x) for x in labels])

        return new_labels



np.random.seed(2)

# Generate more ball-like clusters
n_clusters = 4
n_points_per_cluster = 100
more_cluster_data = []
for _ in range(n_clusters):
    center = np.random.rand(3) * 10  # Random center
    points = np.random.randn(n_points_per_cluster, 3) * 0.2 + center  # Points around the center
    more_cluster_data.append(points)
more_cluster_data = np.vstack(more_cluster_data)

# Generate another noisy line
n_line_points = 100
more_line_points = np.linspace(0, 12, n_line_points)
more_line_data = np.column_stack([more_line_points, more_line_points, more_line_points])
more_line_data += np.random.randn(n_line_points, 3) * 0.2  # Add more noise

# Generate another curve (sinusoidal)
theta = np.linspace(0, 4 * np.pi, 100)
x = np.cos(theta)
y = np.sin(2 * theta)
z = theta / 5
more_curve_data = np.column_stack([x, y, z])
more_curve_data *= 3  # Scale
more_curve_data += np.random.randn(*more_curve_data.shape) * 0.4  # Add more noise

# Combine all data
more_combined_data = np.vstack([more_cluster_data, more_line_data, more_curve_data])



# Initialize the clustering model with the new class
clustering_model = ImprovedTemporalSpatialClusterWithNN(density_threshold=1.0,
                                                        temporal_threshold=10.0,
                                                        n_neighbors=3,
                                                        angle_threshold=0.8)
X = more_combined_data
# Fit the model and get the labels
labels = clustering_model.fit_predict(X)
new_labels = clustering_model.merge_clusters(X, labels, merge_threshold=3.0)

unique_labels = np.unique(new_labels)
print(f"Number of unique labels (clusters): {len(unique_labels)}")

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(more_combined_data[:, 0], more_combined_data[:, 1], more_combined_data[:, 2], c=new_labels, cmap='viridis')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# rd.seed(0)

class Function(object):
    # 
    def sigmoid(self, x):
        return 1. / (1 + np.exp(-x))
    # 
    def dsigmoid(self, x):
        return x * (1. - x)
    # 
    def tanh(self, x):
        return np.tanh(x)
    # 
    def dtanh(self, x):
        return 1. - x * x
    # 
    def ReLU(self, x):
        return x * (x > 0)
    # 
    def dReLU(self, x):
        return 1. * (x > 0)
    # 
    def softmax(self, x):
        e = np.exp(x - np.max(x))  # prevent overflow
        if e.ndim == 1:
            return e / np.sum(e, axis=0)
        else:
            return e / np.array([np.sum(e, axis=1)]).T  # ndim = 2

class Newral_Network(object):
    def __init__(self, unit):
        print("Number of layer = ",len(unit))
        print(unit)
        print("-------------------------")
        self.F = Function()
        self.unit = unit
        self.W = []
        self.B = []
        self.dW = []

        for i in range(len(self.unit) - 1):
            w = np.random.rand(self.unit[i], self.unit[i + 1])
            self.W.append(w)
            # ,
            dw = np.random.rand(self.unit[i], self.unit[i + 1])
            self.dW.append(dw)
            b = np.random.rand(self.unit[i + 1])
            self.B.append(b)

    # 
    def U(self, x, w, b):
        return np.dot(x, w) + b

    # 
    def calc_grad(self, w, b, z, delta):
        w_grad = np.zeros_like(w)
        b_grad = np.zeros_like(b)
        N = float(z.shape[0])
        w_grad = np.dot(z.T, delta) / N
        b_grad = np.mean(delta, axis=0)
        return w_grad, b_grad

    # 
    def backPropagate(self, _label, eta, M):
        # calculate output_delta and error terms
        W_grad = []
        B_grad = []
        for i in range(len(self.W)):
            w_grad = np.zeros_like(self.W[i])
            W_grad.append(w_grad)
            b_grad = np.zeros_like(self.W[i])
            B_grad.append(b_grad)
        output = True
        delta = np.zeros_like(self.Z[-1])
        for i in range(len(self.W)):
            delta = self.calc_delta(
                delta, self.W[-(i)], self.Z[-(i + 1)], _label, output)

            W_grad[-(i + 1)], B_grad[-(i + 1)] \
                = self.calc_grad(self.W[-(i + 1)], self.B[-(i + 1)], self.Z[-(i + 2)], delta)

            output = False

        # update weights
        for i in range(len(self.W)):
            self.W[i] = self.W[i] - eta * W_grad[i] + M * self.dW[i]
            self.B[i] = self.B[i] - eta * B_grad[i]
            # 
            self.dW[i] = -eta * W_grad[i] + M * self.dW[i]
    # 
    def train(self, dataset, N, iterations=1000, minibatch=4, eta=0.5, M=0.1):
        print("-----Train-----")
        # 
        inputs = dataset[:, :self.unit[0]]

        # 
        label = dataset[:, self.unit[0]:]

        errors = []
        for val in range(iterations):
            minibatch_errors = []
            for index in range(0, N, minibatch):
                _inputs = inputs[index: index + minibatch]
                _label = label[index: index + minibatch]
                self.forward(_inputs)
                self.backPropagate(_label, eta, M)

                loss = self.calc_loss(_label)
                minibatch_errors.append(loss)
            En = sum(minibatch_errors) / len(minibatch_errors)
            print("epoch", val + 1, " : Loss = ", En)
            errors.append(En)
        print("\n")
        errors = np.asarray(errors)
        #plt.plot(errors)

    # 
    def getWeights(self):
        weights = []
        biases = []
        for i in range(len(self.W)):
            weights.append(self.W[i])
            biases.append(self.B[i])
        return weights, biases


    # 
    def save_weight(self, name):
        import datetime
        today_detail = datetime.datetime.today()
        s = today_detail.strftime("%m-%d-%H-%M")
        np.save('Models/%s_W_%s.npy' % (name, s), self.W)
        np.save('Models/%s_B_%s.npy' % (name, s), self.B)
        print("Weight is saved!!")
        for i in range(len(self.W)):
            print("W", i + 1, ".shape = ", self.W[i].shape)
            print ("B", i + 1, ".shape = ", self.B[i].shape)
        print("\n")

    def forward(self, _inputs):
      self.Z = []
      self.Z.append(_inputs)
      for i in range(len(self.unit) - 1):
          u = self.U(self.Z[i], self.W[i], self.B[i])
          if(i != len(self.unit) - 2):
              z = np.tanh(u)
          else:
              z = self.F.softmax(u)
          self.Z.append(z)

      return z  # 

    # 
    def calc_loss(self, label):
        error = np.sum(label * np.log(self.Z[-1]), axis=1)
        return -np.mean(error)

    # 
    def calc_delta(self, delta_dash, w, z, label, output):
        # delta_dash : 1
        # w : pre_deltadelta
        # z : w
        if(output):
            delta = z - label
        else:
            delta = np.dot(delta_dash, w.T) * self.F.dtanh(z)
        return delta

    def draw_test(self, x, label, W, B):
        self.W = W
        self.B = B
        x1_max = max(x[:, 0]) + 0.5
        x2_max = max(x[:, 1]) + 0.5
        x1_min = min(x[:, 0]) - 0.5
        x2_min = min(x[:, 1]) - 0.5
        xx, yy = np.meshgrid(np.arange(x1_min, x1_max, 0.01),
                             np.arange(x2_min, x2_max, 0.01))
        Z = self.forward(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)

        #plt.contour(xx, yy, Z)
        plt.contourf(xx, yy, Z, cmap=plt.cm.jet)
        plt.scatter(x[:, 0], x[:, 1], c=label, cmap=plt.cm.jet)
        plt.show()

    def test(self, x, W, B):
      self.W = W
      self.B = B
      print("-----Test-----")
      for i in range(x.shape[0]):
          print("input = ", x[i])
          output_prob = self.forward(x[i])  # 
          # 
          if output_prob.ndim == 1:
              output_class = int(np.argmax(output_prob))
          else:
              output_class = int(np.argmax(output_prob, axis=1))
          print("output = ", output_class)
          print("\n")

    def draw_test1(self, x, label, W, B):
        self.W = W
        self.B = B
        x1_max = max(x[:, 0]) + 0.5
        x2_max = max(x[:, 1]) + 0.5
        x1_min = min(x[:, 0]) - 0.5
        x2_min = min(x[:, 1]) - 0.5
        xx, yy = np.meshgrid(np.arange(x1_min, x1_max, 0.01),
                         np.arange(x2_min, x2_max, 0.01))
        Z = self.forward(np.c_[xx.ravel(), yy.ravel()])
        # 
        # np.argmax 
        if Z.shape[1] > 1:
            Z = np.argmax(Z, axis=1)  # 
        else:
            Z = Z.squeeze()  # 

        Z = Z.reshape(xx.shape)  # 

        plt.contourf(xx, yy, Z, cmap=plt.cm.jet)
        plt.scatter(x[:, 0], x[:, 1], c=label, cmap=plt.cm.jet)
        plt.show()
```

```python
# 
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
from collections import OrderedDict

class AutoEncoder(nn.Module):
    def __init__(self, args):
        super(AutoEncoder, self).__init__()
        self.args = args
        self.input_dim = args.input_dim
        self.hidden_dims = args.hidden_dims
        self.latent_dim = args.latent_dim
        # Ensure the hidden dimensions are set up for a symmetric encoder/decoder
        self.hidden_dims.append(self.latent_dim)
        self.hidden_dims = self.hidden_dims + self.hidden_dims[:-1][::-1]
        self.output_dim = self.input_dim

        # Encoder layers
        encoder_layers = OrderedDict()
        in_dim = self.input_dim
        for i, hidden_dim in enumerate(self.hidden_dims):
            encoder_layers[f'linear_{i}'] = nn.Linear(in_dim, hidden_dim)
            if i < len(self.hidden_dims) - 1:  # no ReLU/BatchNorm in the bottleneck layer
                encoder_layers[f'relu_{i}'] = nn.ReLU()
                encoder_layers[f'bn_{i}'] = nn.BatchNorm1d(hidden_dim)
            in_dim = hidden_dim
        self.encoder = nn.Sequential(encoder_layers)

        # Decoder layers
        decoder_layers = OrderedDict()
        self.hidden_dims.reverse()  # Reverse the hidden dims for decoder
        for i in range(len(self.hidden_dims) - 1):
            decoder_layers[f'linear_{len(self.hidden_dims) + i}'] = nn.Linear(self.hidden_dims[i], self.hidden_dims[i + 1])
            if i < len(self.hidden_dims) - 2:  # no ReLU/BatchNorm in the last layer
                decoder_layers[f'relu_{len(self.hidden_dims) + i}'] = nn.ReLU()
                decoder_layers[f'bn_{len(self.hidden_dims) + i}'] = nn.BatchNorm1d(self.hidden_dims[i + 1])
        decoder_layers[f'linear_{len(self.hidden_dims) * 2 - 2}'] = nn.Linear(self.hidden_dims[-2], self.output_dim)
        self.decoder = nn.Sequential(decoder_layers)

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x


# 
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# MNIST
train_dataset = datasets.MNIST('./data', train=True, download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)

# Args
class Args:
    def __init__(self):
        self.input_dim = 784  # 28x28 images flattened
        self.hidden_dims = [128, 64]
        self.latent_dim = 32
        self.n_clusters = 10  # For example, if using MNIST

args = Args()

# AutoEncoder
model = AutoEncoder(args)

# 
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 
num_epochs = 10
for epoch in range(num_epochs):
    for data in train_loader:
        inputs, _ = data
        optimizer.zero_grad()
        outputs = model(inputs.view(inputs.shape[0], -1))
        loss = criterion(outputs, inputs.view(inputs.shape[0], -1))
        loss.backward()
        optimizer.step()
    print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))

# 
dataiter = iter(train_loader)
images, labels = next(dataiter)
output = model(images.view(images.shape[0], -1))
images = images.numpy()

output = output.view(output.size(0), 1, 28, 28).detach().numpy()

fig, axes = plt.subplots(nrows=2, ncols=10, sharex=True, sharey=True, figsize=(25,4))

# Original images and reconstructed images
for images, row in zip([images, output], axes):
    for img, ax in zip(images, row):
        ax.imshow(np.squeeze(img), cmap='gray')
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
plt.show()
```

```python
# 
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
from collections import OrderedDict

class AutoEncoder(nn.Module):
    def __init__(self, args):
        super(AutoEncoder, self).__init__()
        self.args = args
        self.input_dim = args.input_dim
        self.hidden_dims = args.hidden_dims
        self.latent_dim = args.latent_dim
        # Ensure the hidden dimensions are set up for a symmetric encoder/decoder
        self.hidden_dims.append(self.latent_dim)
        self.hidden_dims = self.hidden_dims + self.hidden_dims[:-1][::-1]
        self.output_dim = self.input_dim

        # Encoder layers
        encoder_layers = OrderedDict()
        in_dim = self.input_dim
        for i, hidden_dim in enumerate(self.hidden_dims):
            encoder_layers[f'linear_{i}'] = nn.Linear(in_dim, hidden_dim)
            if i < len(self.hidden_dims) - 1:  # no ReLU/BatchNorm in the bottleneck layer
                encoder_layers[f'relu_{i}'] = nn.ReLU()
                encoder_layers[f'bn_{i}'] = nn.BatchNorm1d(hidden_dim)
            in_dim = hidden_dim
        self.encoder = nn.Sequential(encoder_layers)

        # Decoder layers
        decoder_layers = OrderedDict()
        self.hidden_dims.reverse()  # Reverse the hidden dims for decoder
        for i in range(len(self.hidden_dims) - 1):
            decoder_layers[f'linear_{len(self.hidden_dims) + i}'] = nn.Linear(self.hidden_dims[i], self.hidden_dims[i + 1])
            if i < len(self.hidden_dims) - 2:  # no ReLU/BatchNorm in the last layer
                decoder_layers[f'relu_{len(self.hidden_dims) + i}'] = nn.ReLU()
                decoder_layers[f'bn_{len(self.hidden_dims) + i}'] = nn.BatchNorm1d(self.hidden_dims[i + 1])
        decoder_layers[f'linear_{len(self.hidden_dims) * 2 - 2}'] = nn.Linear(self.hidden_dims[-2], self.output_dim)
        self.decoder = nn.Sequential(decoder_layers)

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x


    def encode(self, x):
        """
        This function applies the encoder part of the AutoEncoder to
        input data x, returning the encoded (latent) representation.
        """
        return self.encoder(x)

    def forward(self, x):
        encoded = self.encode(x)
        decoded = self.decoder(encoded)
        return decoded

args = Args()
args.input_dim = 300  # N
args.hidden_dims = [50, 25]  # 
args.latent_dim = 10  # 

# 
model = AutoEncoder(args)

# 
# ...

# 10064
input_data = torch.randn(64, args.input_dim)

# 
features = model.encode(input_data)

print(features.shape)  # (64, args.latent_dim)

model = AutoEncoder(args)

# 3*N64
# N=100
N = 100
input_data = torch.randn(64, 3 * N)

# 01
scaler = StandardScaler()
input_data_standardized = scaler.fit_transform(input_data)

# Tensor
input_data_numpy = np.array(input_data_standardized)
input_data_tensor = torch.FloatTensor(input_data_numpy)

# 
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 
num_epochs = 1000
for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = model(input_data_tensor)
    loss = criterion(outputs, input_data_tensor)
    loss.backward()
    optimizer.step()
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# 
with torch.no_grad():  # 
    encoded_features = model.encoder(input_data_tensor)

# PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
encoded_features_2d = pca.fit_transform(encoded_features)

# 
plt.figure(figsize=(8, 6))
plt.scatter(encoded_features_2d[:, 0], encoded_features_2d[:, 1], c='blue', alpha=0.5)
plt.title('Encoded Feature Visualization')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

```python
# 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf, pacf

# 
np.random.seed(0)  # 
data = np.random.rand(100, 3)  # 100x3
columns = ['frequency_centroid', 'bandwidth_variance', 'spectral_flatness']
df = pd.DataFrame(data, columns=columns)

# 
window_size = 5
rolling_means = df.rolling(window=window_size).mean()
rolling_std = df.rolling(window=window_size).std()

# 
lag_features = 3
for column in df.columns:
    for lag in range(1, lag_features + 1):
        df[f'{column}_lag_{lag}'] = df[column].shift(lag)

# 
lag_acf_pacf = 20
acf_values = {col: acf(df[col].dropna(), nlags=lag_acf_pacf) for col in df.columns}
pacf_values = {col: pacf(df[col].dropna(), nlags=lag_acf_pacf) for col in df.columns}

# 
plt.figure(figsize=(12, 8))

# 
plt.subplot(3, 1, 1)
plt.plot(rolling_means, label='Rolling Mean')
plt.title('Rolling Mean')
plt.legend()

# 
plt.subplot(3, 1, 2)
plt.plot(rolling_std, label='Rolling Std Dev')
plt.title('Rolling Standard Deviation')
plt.legend()

# ACFPACF
plt.subplot(3, 1, 3)
for col in df.columns:
    plt.plot(acf_values[col], marker='o', label=f'ACF {col}')
    plt.plot(pacf_values[col], marker='x', label=f'PACF {col}')
plt.title('Auto-correlation and Partial Auto-correlation')
plt.legend()

plt.tight_layout()
plt.show()
```
